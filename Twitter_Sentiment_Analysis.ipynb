```python
#1. Download libraries
import nltk
nltk.download('punkt_tab')
nltk.download('brown')

import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# 2. Twitter Data (College file ke liye clean data)
tweets_data = {
    'Tweets': [
        "I love this product, it is absolutely amazing!",
        "This is the worst service ever, so disappointed.",
        "The weather is okay today, nothing special.",
        "Great experience! Highly recommend it to everyone.",
        "Horrible app, it keeps crashing every minute.",
        "I am flying to Delhi tomorrow."
    ]
}

df = pd.DataFrame(tweets_data)

# 3. Sentiment Pata Karne Ka Function (Positive, Negative, Neutral)
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Tweets par function apply karo
df['Result'] = df['Tweets'].apply(get_sentiment)

# Output print karo
print("--- Tweets and Their Sentiments ---")
print(df)
print("\n-----------------------------------")

# 4. Pie Chart (Gol Graph) Banakar Dikhana
sentiment_counts = df['Result'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['green', 'red', 'blue'])
plt.title('Twitter Sentiment Analysis Results')
plt.show()
```
